Fast-Food Calorie Dataset Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
fastfood_calories = pd.read_csv(r"C:\Users\DELL\Downloads\fastfood_calories.csv")
```

```python
column = fastfood_calories['calories']
```

```python
mean_value = column.mean()
median_value = column.median()
std_deviation = column.std()
```

```python
mean_value, median_value, std_deviation
```

```python
nutritional_factors = ['cal_fat', 'total_fat', 'sat_fat','trans_fat','cholesterol','fiber','sugar','protein','vit_a','vit_c','calcium']
mean_nutritional_factors = fastfood_calories.groupby(['calories'])[nutritional_factors].mean()
```

```python
mean_nutritional_factors.plot(kind='bar',
                              xlabel='mean_value',
                              ylabel='calories',
                              title='Mean Nutritional Factors by Calorie Range',
                              )
plt.show()
```
