```python
order_file = pd.read_excel(r"C:\Users\DELL\Downloads\order file.xlsx")
items_correlation = pd.read_excel(r"C:\Users\DELL\Downloads\items-correlation.xlsx")
rca = pd.read_excel(r"C:\Users\DELL\Downloads\RCA.xlsx")
```

```python
def get_status(row):
    if row['Delivered Qty'] < row['Ordered Qty']:
        return 'Gap in Orders'
    else:
        return 'Delivered'

order_file['order_status'] = order_file.apply(get_status, axis=1)
```

```python
order_file = pd.merge(order_file, items_correlation, on='Items',how='left')
```

```python
def get_rca_description(row):
    if pd.isnull(row['RCA Id']):
        return 'No RCA Found'
    else:
        return 'RCA Found'
```

```python
order_file['RCA Desc.'] = order_file.apply(get_rca_description, axis = 1)
```
