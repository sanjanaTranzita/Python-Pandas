PIVOT FUNCTION

```python
match_up_impact.pivot(index='Bowler', columns='Batter', values='strike_rate')
```

```python
top_scorer_batsman = match_up_impact.groupby(['Batter']).agg({'strike_rate':'mean'}).sort_values(by='strike_rate', ascending=False).head(20)
top_scorer_batsman.plot(kind='bar',
                        xlabel='Batter', 
                        ylabel='Strike rate', 
                        title='Top Performing Batsmen',)

plt.show()
```

```python

top_scorer_batsman = match_up_impact.groupby(['Batter']).agg({'runs_scored':'sum'}).sort_values(by='runs_scored', ascending=False).head(30)
top_scorer_batsman.plot(kind='bar',
                        xlabel='Batter', 
                        ylabel='Total Runs', 
                        title='Top Performing Batsman',)

plt.show()
```
